#2. Data Analysis

```python
# Plots the bar plot of the numerical features
def plot_hist(train_df, variable):
    plt.hist(train_df[variable], bins = 50)
    plt.xlabel(variable)
    plt.ylabel("Frequency")
    plt.title("{} distribution with hist".format(variable))
    plt.show()
    print("\n")

# Plotting histogram of the numerical features 
num_variable = (datasets.dtypes=="float64") | (datasets.dtypes=="int64")
num_variable = datasets.columns[num_variable].tolist()    
for i in num_variable:
    plot_hist(datasets,i)
```

```python
# Plots the bar plot of the categorical features
def bar_plot(data,feature):
    print('There are {0} unique {1}'.format(len(set(data[feature])),feature))
    sns.countplot(x = feature,data = data,order = data[feature].value_counts(ascending=False)[0:20].index)
    plt.xticks(rotation=90)
    plt.show()
    print('\n')
```

```python
# Plotting the categorical features
cat_variable = datasets.dtypes==object
cat_variable = datasets.columns[cat_variable].tolist()

for i in cat_variable:
    bar_plot(datasets,i)
```

#3. Data Cleaning Manipulation

**Removing Outliers from the dataset**

```python

datasets['wheelbase'] = datasets['wheelbase']
datasets['enginesize'] = datasets['enginesize']
datasets['highwaympg'] = datasets['highwaympg']
datasets['price'] = datasets['price']
```

```python
Q1 = datasets.quantile(q=.25)
Q3 = datasets.quantile(q=.75)
IQR = Q3 - Q1
```

```python
# Removing Outliers from the dataset
filtered_datasets =  datasets[~((datasets < (Q1-1.5*IQR)) | (datasets > (Q3+1.5*IQR))).any(axis=1)]
filtered_datasets.dropna()
```

```python
# Plotting numerical features after removing outliers 
fnum_variable = (filtered_datasets.dtypes=="float64") | (filtered_datasets.dtypes=="int64")
fnum_variable = filtered_datasets.columns[fnum_variable].tolist()    
for i in fnum_variable:
  sns.boxplot(x=filtered_datasets[i])
  plt.title("After removing Outliers of {}".format(i))
  plt.show()
  print('\n')
```

```python
# Plotting the heat map of the correlation of the filtered data
sns.heatmap(filtered_datasets.corr(),annot=True)
plt.figure(figsize=(20.0, 10.0)) 
```

#4. Regression task and its accuracy

```python
carname = filtered_datasets['CarName'].values
fueltype = filtered_datasets['fueltype'].values
aspiration = filtered_datasets['aspiration'].values
doornumber = filtered_datasets['doornumber'].values
carbody = filtered_datasets['carbody'].values
drivewheel = filtered_datasets['drivewheel'].values
enginelocation = filtered_datasets['enginelocation'].values
wheelbase = filtered_datasets['wheelbase'].values
carlength = filtered_datasets['carlength'].values
carwidth = filtered_datasets['carwidth'].values
carheight = filtered_datasets['carheight'].values
curbweight = filtered_datasets['curbweight'].values
enginetype = filtered_datasets['enginetype'].values
cylindernumber = filtered_datasets['cylindernumber'].values
enginesize = filtered_datasets['enginesize'].values
fuelsystem = filtered_datasets['fuelsystem'].values
boreratio = filtered_datasets['boreratio'].values
stroke = filtered_datasets['stroke'].values
horsepower = filtered_datasets['horsepower'].values
highwaympg= filtered_datasets['highwaympg'].values
price= filtered_datasets['price'].values

```

```python
from sklearn.linear_model import LinearRegression

# X and Y Values
X = np.array([carname,fueltype,aspiration,doornumber,carbody,drivewheel,enginelocation,wheelbase,carlength,carwidth,carheight,curbweight,enginetype,cylindernumber,enginesize,fuelsystem,boreratio,stroke,horsepower,highwaympg,price]).T
Y = np.array(price)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,Y, test_size=0.9)
```

```python
# Model Intialization
reg = LinearRegression()
# Data Fitting
reg = reg.fit(X_train, y_train)
# Y Prediction
Y_pred = reg.predict(X_test)
```

```python
reg.intercept_
```

```python
pd.DataFrame(reg.coef_,columns = ['Coeff'])
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

# Model Evaluation - RMSE
rmse = np.sqrt(mean_squared_error(y_test, Y_pred))
# Model Evaluation - R2 Score
r2 = r2_score(y_test, Y_pred)

print('Root Mean Square Error is {}'.format(rmse))
print('Coefficient of Determination is {}'.format(r2))
```
